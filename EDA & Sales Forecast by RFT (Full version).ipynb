```python
# prompt: import numpy

import numpy as np
import pandas as pd
```

```python
#2.Check header, row, column (df.head())
df = pd.read_csv("dlbh.csv")
df.head(3)
```

```python
#3.Check info (df.info())
df.info()
```

```python
# prompt: Delete % in Tỷ trọng

# Assuming 'Tỷ trọng' is the column name containing the percentage values.
df['Tỷ trọng'] = df['Tỷ trọng'].str.replace('%', '')
df.head(3)
```

```python
# prompt: Convert "," of Tỷ trọng to "."

df['Tỷ trọng'] = df['Tỷ trọng'].str.replace(',', '.')
df.head(3)
```

```python
# prompt: Transfer data type of Tỷ trọng to float64

df['Tỷ trọng'] = df['Tỷ trọng'].astype('float64')
df.info()
```

```python
# prompt: divide Tỷ trọng by 100. Divide Tổng tiền hàng bán by 1000

# Divide 'Tỷ trọng' by 100
df['Tỷ trọng'] = df['Tỷ trọng'] / 100

# Divide 'Tổng tiền hàng bán' by 1000
df['Tổng tiền hàng bán'] = df['Tổng tiền hàng bán'] / 1000

df.head(3)
```

```python
# prompt: Please convert Year, Month, Day to datetime64
# Convert 'Year', 'Month', and 'Day' columns to datetime64
df['Dates'] = pd.to_datetime(df[['Year', 'Tháng (số)', 'Day']].astype(str).apply(lambda x: '-'.join(x), axis=1), format='%Y-%m-%d')
df['Dates'] = pd.to_datetime(df['Dates'])
```

```python
# prompt: Remove outliers of Tổng tiền hàng bán. Replace the old data with the new version one

# Calculate the IQR (Interquartile Range)
Q1 = df['Tổng tiền hàng bán'].quantile(0.25)
Q3 = df['Tổng tiền hàng bán'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_filtered = df[
    (df['Tổng tiền hàng bán'] >= lower_bound) &
    (df['Tổng tiền hàng bán'] <= upper_bound)
]

# Replace the original DataFrame with the filtered one
df = df_filtered
```

```python
# prompt: Correlation matrix for Chiết khấu, Lợi nhuận, Giá bán, Giá vốn hàng bán, Giá xuất xưởng, Tổng tiền hàng bán

# Select the columns for correlation analysis
selected_columns = ['Chiết khấu', 'Lợi nhuận', 'Giá bán', 'Giá vốn hàng bán', 'Giá xuất xưởng', 'Tổng tiền hàng bán']

# Calculate the correlation matrix for the selected columns
corr_matrix = df[selected_columns].corr()

# Display the correlation matrix
print(corr_matrix)
```

```python
# prompt: Linear regression for a scatter plot of 'Tổng tiền hàng bán' against 'Lợi nhuận'

import matplotlib.pyplot as plt
import statsmodels.api as sm

# Add a constant term to the independent variable
X = sm.add_constant(df['Tổng tiền hàng bán'])
y = df['Lợi nhuận']

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Get the predicted values
predictions = model.predict(X)

# Plot the scatter plot with the regression line
plt.figure(figsize=(4, 3))
plt.scatter(df['Tổng tiền hàng bán'], df['Lợi nhuận'])
plt.plot(df['Tổng tiền hàng bán'], predictions, color='red')
plt.xlabel('Tổng tiền hàng bán')
plt.ylabel('Lợi nhuận')
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
# prompt: which months has the highest and lowest Tổng tiền hàng bán in each Year. How much

import pandas as pd
# Group by 'Year' and 'Tháng (số)' and sum 'Tổng tiền hàng bán'
monthly_sales = df.groupby(['Year', 'Tháng (số)'])['Tổng tiền hàng bán'].sum().reset_index()

# Find the month with the highest and lowest sales for each year
def find_highest_lowest(group):
  highest = group.loc[group['Tổng tiền hàng bán'].idxmax()]
  lowest = group.loc[group['Tổng tiền hàng bán'].idxmin()]
  return pd.DataFrame({'Highest': [highest], 'Lowest': [lowest]})

# Apply the function to find the highest and lowest sales months for each year
result = monthly_sales.groupby('Year').apply(find_highest_lowest).reset_index()

# Extract the highest and lowest sales months and their values
highest_sales_months = result['Highest'].apply(pd.Series)
lowest_sales_months = result['Lowest'].apply(pd.Series)

# Display the results
print("Highest Sales Months:")
print(highest_sales_months[['Year', 'Tháng (số)', 'Tổng tiền hàng bán']])
print("\nLowest Sales Months:")
print(lowest_sales_months[['Year', 'Tháng (số)', 'Tổng tiền hàng bán']])
```

```python
# prompt: Do RFM analysis with 'Khách hàng' as a customer name, 'Loại khách hàng' as a customer type, 'Tổng tiền hàng bán' as a monetary value, and the times 'Khách hàng' buy as a frequency. Show top 20 'Khách hàng' has the highest RFM score and top 20 lowest RFM score (Order by Monetary value DESC). How much?

import pandas as pd
# Calculate Recency
df['Dates'] = pd.to_datetime(df['Dates'])
recent_date = df['Dates'].max()
rfm_df = df.groupby('Khách hàng').agg(
    Recency=('Dates', lambda x: (recent_date - x.max()).days),
    Frequency=('Khách hàng', 'count'),
    Monetary_Value=('Tổng tiền hàng bán', 'sum')
)

# Calculate RFM Score
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm_df['Monetary_Value_Score'] = pd.qcut(rfm_df['Monetary_Value'], 5, labels=[1, 2, 3, 4, 5])

rfm_df['RFM_Score'] = rfm_df['Recency_Score'].astype(int) + rfm_df['Frequency_Score'].astype(int) + rfm_df['Monetary_Value_Score'].astype(int)

# Sort by RFM Score and Monetary Value
rfm_df_sorted = rfm_df.sort_values(['RFM_Score', 'Monetary_Value'], ascending=[False, False])

# Top 20 highest RFM score
top_20_highest = rfm_df_sorted.head(20)
print("Top 20 highest RFM score:")
print(top_20_highest)

# Top 20 lowest RFM score
top_20_lowest = rfm_df_sorted.tail(20)
print("\nTop 20 lowest RFM score:")
print(top_20_lowest)
```

```python
# prompt: Which Khách hàng has highest monetary value. How much? Also show its Recency, Frequency and RFM_Score

# Find the Khách hàng with the highest Monetary Value
highest_monetary_value_customer = rfm_df_sorted.loc[rfm_df_sorted['Monetary_Value'].idxmax()]

# Print the results
print("Khách hàng with the highest Monetary Value:", highest_monetary_value_customer.name)
print("Monetary Value:", highest_monetary_value_customer['Monetary_Value'])
print("Recency:", highest_monetary_value_customer['Recency'])
print("Frequency:", highest_monetary_value_customer['Frequency'])
print("RFM_Score:", highest_monetary_value_customer['RFM_Score'])
```

```python
# prompt: Use LSTM model to predict each 'Chiết khấu', 'Giá bán', 'Giá vốn hàng bán', 'Giá xuất xưởng', 'Số lượng sản phẩm' for the next 30 Days (based on 'Dates'). Store those predicted values to an array named as 'random_X'. Visualize 'random_X' with future 'Dates'.

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Select the columns for prediction
columns_to_predict = ['Chiết khấu', 'Giá bán', 'Giá vốn hàng bán', 'Giá xuất xưởng', 'Số lượng sản phẩm']

# Prepare the data for LSTM
data = df[['Dates'] + columns_to_predict].copy()
data['Dates'] = pd.to_datetime(data['Dates'])
data.set_index('Dates', inplace=True)

# Scale the data
scaler = MinMaxScaler()
data[columns_to_predict] = scaler.fit_transform(data[columns_to_predict])

# Split the data into training and testing sets
train_data = data[:-30]
test_data = data[-30:]

# Create the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(train_data.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(len(columns_to_predict)))
model.compile(loss='mean_squared_error', optimizer='adam')

# Prepare the data for the LSTM model
X_train = np.array(train_data).reshape(train_data.shape[0], train_data.shape[1], 1)
y_train = np.array(train_data)

# Train the LSTM model
model.fit(X_train, y_train, epochs=500, batch_size=64)

# Prepare the data for prediction
X_test = np.array(test_data).reshape(test_data.shape[0], test_data.shape[1], 1)

# Predict the values
predicted_values = model.predict(X_test)

# Inverse transform the predicted values
predicted_values = scaler.inverse_transform(predicted_values)

# Create the future dates
future_dates = pd.date_range(start=df['Dates'].max(), periods=30, freq='D')

# Store the predicted values in random_X
random_X = pd.DataFrame(predicted_values, columns=columns_to_predict, index=future_dates)

# Visualize the predictions
plt.figure(figsize=(12, 6))
for column in columns_to_predict:
  plt.plot(random_X.index, random_X[column], label=column)
plt.xlabel('Dates')
plt.ylabel('Values')
plt.title('Predicted Values for Next 30 Days')
plt.legend()
plt.show()
```

```python
# prompt: Use random forest model to predict Tổng tiền hàng bán based on 'Chiết khấu', 'Giá bán', 'Giá vốn hàng bán', 'Giá xuất xưởng', 'Số lượng sản phẩm' by the next 30 days (based on 'Date') with 'random_X' as a sample for the next 30 days prediction.

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Select features and target variable
features = ['Chiết khấu', 'Giá bán', 'Giá vốn hàng bán', 'Giá xuất xưởng', 'Số lượng sản phẩm']
target = 'Tổng tiền hàng bán'

n = df[features]
m = df[target]

# Split data into training and testing sets
n_train, n_test, m_train, m_test = train_test_split(n, m, test_size=0.2, random_state=42)

# Train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(n_train, m_train)

# Make predictions on the test set
m_pred = rf_model.predict(n_test)

# Evaluate the model
mse = mean_squared_error(m_test, m_pred)
print(f"Mean Squared Error: {mse}")

# Predict 'Tổng tiền hàng bán' for the next 30 days
random_X['Tổng tiền hàng bán'] = rf_model.predict(random_X[features])

# Print the predicted values for the next 30 days
print(random_X[['Tổng tiền hàng bán']])
```
