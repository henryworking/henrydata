```python
#3.1:
a = int(input('Nhập 1 số nguyên a (>= 0): '))
b = int(input('Nhập 1 số nguyên b (>= 0 và a<b): '))
snt = []

if a < b:
  for i in range(a, b+1):
    count1 = 0
    for j in range(1,i):    # kiểm tra số chia hết cho 1 và chính nó
      if i % j ==0:
        count1 = count1 + 1
    if count1 == 1:
      snt.append(i)
  print('Danh sách số nguyên tố là:', snt)
else:
  print('Sai định dạng dữ liệu, số A phải nhỏ hơn B. Vui lòng nhập lại')
```

```python
# 3.2(1đ). Viết function game số may mắn. Nội dung in ra màn hình để người dùng nhập vào số tiền muốn cược, số muốn cược(00-99). 
# Random 1 số có 2 chữ số và in ra màn hình số đó. Nếu số random trùng với số đặt cược, in ra màn hình. Bạn đã trúng thưởng. 
# Số tiền thưởng bạn nhận được là tiền đặt cược*80. Nếu số random khác số may mắn thì in ra là bạn không trúng thưởng. 
# Sau đó hỏi người chơi muốn tiếp tục không(y/n)? Nếu người chơi chọn yes thì lặp lại nội dung trên. 
#Nếu người chơi chọn no thì kết thúc vòng lặp, in ra màn hình “Cảm ơn bạn đã tham gia, hẹn gặp bạn lần sau”/

while True:
  import random
  lucky = random.randint(0,99)
  #print('Số may mắn là:',lucky)    # in ra số may mắn trước để check code cho nhanh :)))
  money = float(input('Chào mừng bạn đến với Game số may mắn. Hãy nhập vào số tiền bạn muốn đặt cược (đvt: triệu đồng):'))
  numb = int(input('Hãy nhập vào số bạn muốn đặt cược (0-99): '))

  if numb != lucky:
    print('Bạn không trúng thưởng. Chúc may mắn lần sau!! Số may mắn là:', lucky)
    cont = str(input('Bạn có muốn thử lại vận may của mình không (y/n)?'))
    if cont == 'n':
      print('Cảm ơn bạn đã tham gia, hẹn gặp lại lần sau')
      break
    elif cont == 'y':
      continue
  else:  
    lucky_money = money * 80
    print('Xin chúc mừng, bạn đã trúng thưởng')
    print('Số tiền thưởng bạn nhận được là:', lucky_money, 'triệu đồng')
    hoi = str(input('Bạn có muốn thử tiếp tục vận may của mình không (y/n)?'))
    if hoi == 'n':
      print('Cảm ơn bạn đã tham gia, hẹn gặp lại lần sau')
      break
    elif hoi == 'y':
      continue
```
